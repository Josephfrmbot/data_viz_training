```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3])
y = np.array([4,5,6])

plt.plot(x,y, label= "the new beginning",color="hotpink", lw= 5, ls="--")
```

```python
plt.figure()
plt.plot(x,y, label="quadratique")
plt.plot(x,y**3, label="cubique")
plt.title("Ce que je n'aime pas")
plt.xlabel("Tiens tiens tiens")
plt.ylabel("Trop fort")
plt.legend()
plt.savefig('essai1')
```
