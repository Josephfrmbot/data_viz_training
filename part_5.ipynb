# Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Revenue")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Sales")

plt.show()
```

# Scatter

```python
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([15, 32, 55, 30, 42, 19, 11])

colors = np.array(["blue", "green", "purple", "brown", "black", "grey", "yellow"])

plt.scatter(x,y , c= colors, cmap='Spectral_r')
plt.colorbar()

plt.show()
```
