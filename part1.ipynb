# Import librairy

```python
from matplotlib import pyplot as plt
import numpy as np
```

# Drawing a line


```python
xpoints= np.array([0, 100])
ypoints= np.array([0, 25])

plt.plot(xpoints, ypoints)
plt.show()
```

# Draw a line in a diagram from position (11, 33) to position (8, 10):

```python
xpoints = np.array([11, 8])
ypoints = np.array([33, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

# Plotting whitout line

```python
xpoints = np.array([11, 8])
ypoints = np.array([33, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```
