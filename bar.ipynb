# import library

```python
import matplotlib.pyplot as plt
import numpy as np
```

# The points

```python
x= np.array(["1","2","3","4"])
y= np.array([5, 9, 11, 6])

plt.bar(x, y)
plt.show()
```

# The horizontal bars

```python
x= np.array(["1","2","3","4"])
y= np.array([5, 9, 11, 6])

plt.barh(x, y)
plt.show()
```

# The color

```python
x= np.array(["1","2","3","4"])
y= np.array([5, 9, 11, 6])

plt.bar(x,y, color = "red")
plt.show()
```

# The Width

```python
x= np.array(["1","2","3","4"])
y= np.array([5, 9, 11, 6])

plt.bar(x, y, color = "black", width=0.3)
plt.show()
```
