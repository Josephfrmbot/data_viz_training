```python
ypoints = np.array([3, 9, 5, 11])
plt.plot(ypoints, color = 'r')
plt.show()
```

# Make it green


```python
ypoints = np.array([3, 9, 5, 11])
plt.plot(ypoints, color = 'g')
plt.show()
```
