 # Import library 

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Checking the plot 

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([230, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)
plt.xlabel('Average Puls')
plt.ylabel('Calorie Burnage')

plt.show()
```

# Create a title to

```python

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([230, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)

plt.title('Super Data Sports')
plt.xlabel('Average Puls')
plt.ylabel('Calorie Burnage')

plt.show()
```

# Changing fonts

```python

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([230, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)

font1 = {'family':'sans serif','color':'darkblue','size':20}
font2 = {'family':'sans serif','color':'purple','size':15}

plt.title('Super Data Sports',fontdict = font1)
plt.xlabel('Average Pulse2',fontdict = font2)
plt.ylabel('Calorie Burnag2',fontdict = font2)

plt.show()
```

# Positionning the title 

```python

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([230, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)

plt.title('Super Data Sports',loc='right')
plt.xlabel('Average Puls')
plt.ylabel('Calorie Burnage')

plt.show()
```

# Add Grid

```python

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([230, 250, 260, 270, 280, 290, 300, 310, 320, 330])



font1 = {'family':'sans serif','color':'darkblue','size':20}
font2 = {'family':'sans serif','color':'purple','size':15}

plt.title('Super Data Sports',fontdict = font1)
plt.xlabel('Average Pulse2',fontdict = font2)
plt.ylabel('Calorie Burnag2',fontdict = font2)

plt.plot(x,y)
plt.grid()
plt.show()
```

# Specify the y grid

```python

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([230, 250, 260, 270, 280, 290, 300, 310, 320, 330])



font1 = {'family':'sans serif','color':'darkblue','size':20}
font2 = {'family':'sans serif','color':'purple','size':15}

plt.title('Super Data Sports',fontdict = font1)
plt.xlabel('Average Pulse2',fontdict = font2)
plt.ylabel('Calorie Burnag2',fontdict = font2)

plt.plot(x,y)
plt.grid(axis='y')
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([230, 250, 260, 270, 280, 290, 300, 310, 320, 330])



font1 = {'family':'sans serif','color':'darkblue','size':20}
font2 = {'family':'sans serif','color':'purple','size':15}

plt.title('Super Data Sports',fontdict = font1)
plt.xlabel('Average Pulse2',fontdict = font2)
plt.ylabel('Calorie Burnag2',fontdict = font2)

plt.plot(x,y)
plt.grid(axis='y', color='red', linestyle='-')
plt.show()
```
