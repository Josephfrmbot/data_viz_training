# Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ypoints = np.array([3, 9, 5, 11])

plt.plot(ypoints, marker = '*')
plt.show()
```

# Make some circles instead

```python
ypoints = np.array([3, 9, 5, 11])

plt.plot(ypoints, marker = 'o')
plt.show()
```

# Make square

```python
ypoints = np.array([3, 9, 5, 11])

plt.plot(ypoints, marker = 'd')
plt.show()
```

# Create a green dotted line

```python
ypoints = np.array([3, 9, 5, 11])

plt.plot(ypoints, 'o:g')
plt.show()
```

# Change the size of the markers and color

```python
ypoints = np.array([3, 9, 5, 11])

plt.plot(ypoints, 'o:y', ms=22)
plt.show()
```

# Set a marker color

```python
ypoints = np.array([3, 9, 5, 11])

plt.plot(ypoints, 'o:y', ms=22, mec= 'b')
plt.show()
```
